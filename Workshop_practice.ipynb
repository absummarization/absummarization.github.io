####LIST -Find an Index of a value

1. Index methon in a list (ONLY Can find the first index)

```python
#1. Index methon in a list (ONLY Can find the first index)
a = [2,4,7,9,10,11,2,3,10]
b = a.index(10)
print(b)
```

2 iterative method to find the position (ONLY find the first index)

```python
# 2 iterative method to find the position (ONLY find the first index)
a = [2,4,7,9,10,11,2,3,10]

def find(val):

  for i in range(len(a)):
    if a[i]== val:
      return i

print(find(10))
```

3. Enumeration - can find all the indexes in a list (Can find all the indexes of the value) ENUMeration 


```python
# 3. can find all the indexes in a list (Can find all the indexes of the value)

a = [2,4,7,9,10,11,2,3,10]

index_list = []

def find_enum (val):

    for i, x in enumerate(a):
        if x == val:
          index_list.append(i)

    return index_list

print(find_enum(2))


#enumerate(list, start=1) => the index of the list will start from 1 rather than 0
```

####Sorting List

```python
#Sort a list using Sort()

a = [2,4,7,9,10,11,2,3,10]
a.sort()
print(a)

#sort a lit with sort() function and Reverse

b = [2,4,7,9,10,11,2,3,10]
b.sort(reverse=True)
print(b)

#sort a list without changing the Original list itself using sorted(list) function
c = [2,4,7,9,10,11,2,3,10]

d = sorted(c, reverse=True)

print(d)
```

####Loop a List or Multiple List (for loop, ZIP)

```python
colors = ["red", "green", "black"]
numbers = [1,2,3]

for color, number in zip(colors, numbers):

    print("{} - {}".format(color, number))

```
